# Sync Changes to Repo
Run git commands to add, commit, and push the copied codebase.

```python
import subprocess
import os
os.chdir('/workspaces/Memetrader')
subprocess.run(['git', 'status'], check=True)
```

```python
subprocess.run(['git', 'add', '.'], check=True)
subprocess.run(['git', 'commit', '-m', 'Copy hermes-agent codebase from memetrader-unification branch'], check=True)
subprocess.run(['git', 'push'], check=True)
print('Changes synced to repo successfully.')
```
